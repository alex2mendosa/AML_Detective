```python
# api_perplexity_aml

# similar to tavily
from perplexity import Perplexity
import os
# BadRequestError: Error code: 400 - {'error': {'message': '["At body: Value error, max_results must be between 1 and 20"]', 'type': 'bad_request', 'code': 400}}

client = Perplexity(api_key = os.environ.get("PERPLEXITY_API_KEY"))

params = {
    "max_results":20,
    "max_tokens":50,
    "search_language_filter":["en","ro"],
    "search_recency_filter" : "year",
    # "search_domain_filter": ["-reddit.com", "-pinterest.com", "-quora.com"]

}

search = client.search.create(
    query=[
      "Lukoil Money Laundering"
    ], **params
)
```
